```r
marine_fish = read_html ("https://wildlife-species.canada.ca/species-risk-registry/sar/index/default_e.cfm?stype=species&lng=e&index=1&common=&scientific=&population=&taxid=4&locid=15&desid=0&schid=0&desid2=0&")
```

```r
marine_fish %>% html_nodes ('table')
```

```r
fish_data = marine_fish %>% html_nodes('table') %>% .[[1]] %>% 
html_table(fill=TRUE) %>% filter(Taxon!='AllLichensMossesVascular PlantsMolluscsArthropodsFishes (marine)Fishes (freshwater)AmphibiansReptilesBirdsMammals (marine)Mammals (terrestrial)') %>% filter(Taxon!='* A common name search will search all common names in English and French as well as aliases and former names which are not displayed below.')
names(fish_data) = c("Common_name", "Scientific_name", "Population", "Taxon", "Range", "COSEWIC_status", "On_schedule_1", "SARA_status")
fish_data
```
